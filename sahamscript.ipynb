```python
import numpy as np
import pandas as pd
import yfinance as yf

def pentilsaham(saham):
    saham_pentil = []
    for stock in saham:
        stock = stock + '.JK'
        info = yf.Ticker(stock).info

        kode = stock.replace('.JK', '')
        c = info.get('currentPrice')
        min_val = info.get('fiftyTwoWeekLow')
        max2m = info.get('fiftyDayAverage')
        max6m = info.get('twoHundredDayAverage')
        max1y = info.get('fiftyTwoWeekHigh')

        # Handling None values
        c = c if c is not None else 0
        min_val = min_val if min_val is not None else 0
        max2m = max2m if max2m is not None else 0
        max6m = max6m if max6m is not None else 0
        max1y = max1y if max1y is not None else 0

        c = round(c)
        min_val = round(min_val)
        max2m = round(max2m)
        max6m = round(max6m)
        max1y = round(max1y)

        p2m = round(np.interp(c, [min_val, max2m], [0, 100]))
        p6m = round(np.interp(c, [min_val, max6m], [0, 100]))
        p1y = round(np.interp(c, [min_val, max1y], [0, 100]))

        saham_pentil.append({'kode': kode, 'c': c, 'min': min_val, 'max2m': max2m, 'max6m': max6m, 'max1y': max1y, 'p2m': p2m, 'p6m': p6m, 'p1y': p1y})

    return saham_pentil
```

```python
import tensorflow as tf
import numpy as np

def load_model(file_path):
    """
    Memuat model dari file H5.

    :param file_path: str, path ke file H5 yang menyimpan model
    :return: model yang dimuat
    """
    model = tf.keras.models.load_model(file_path)
    return model

def predict_with_model(model, data):
    """
    Menggunakan model untuk membuat prediksi.

    :param model: model yang telah diload
    :param data: data yang akan diprediksi (harus dalam format yang diterima oleh model)
    :return: prediksi dari model
    """
    predictions = model.predict(data)
    return predictions

# Contoh penggunaan:
if __name__ == "__main__":
    # Memuat model
    model = load_model('RNN100Adamax01.h5')

    # Misalkan kita punya data untuk diprediksi
    # Data harus memiliki bentuk (batch_size, timesteps, features)
    # Model mengharapkan 4 fitur per timestep

    # Data contoh: 1 sampel, 100 timesteps, 1 fitur per timestep
    sample_data = np.random.random((1, 100, 1))

    # Mengubah bentuk data menjadi (batch_size, timesteps, 4 features)
    # Menambahkan 3 fitur tambahan dengan nilai nol
    sample_data_adjusted = np.zeros((sample_data.shape[0], sample_data.shape[1], 4))
    sample_data_adjusted[:, :, 0] = sample_data[:, :, 0]

    # Membuat prediksi
    predictions = predict_with_model(model, sample_data_adjusted)
    print(predictions)
```

```python
def digit(angka):
    if angka >= 10**12:
        return f"{angka / 10**12:.1f} T"
    elif angka >= 10**9:
        return f"{angka / 10**9:.1f} M"
    elif angka >= 10**6:
        return f"{angka / 10**6:.1f} Jt"
    elif angka >= 10**3:
        return f"{angka / 10**3:.1f} Rb"
    elif angka <= -10**12:
        return f"{angka / 10**12:.1f} T"
    elif angka <= -10**9:
        return f"{angka / 10**9:.1f} M"
    elif angka <= -10**6:
        return f"{angka / 10**6:.1f} Jt"
    elif angka <= -10**3:
        return f"{angka / 10**3:.1f} Rb"
    else:
        return str(angka)
```

```python
import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

def stand_ruler_chart(value, max_value, title=''):
    # Plotting the stand ruler chart
    fig, ax = plt.subplots(figsize=(6, 3))

    # Plotting the ruler
    ax.plot([0, 1], [0, 0], color='black', lw=10)

    # Plotting the value indicator
    ax.plot(value / max_value, 0, 'ro', markersize=10)

    # Setting the axis limits and hiding the axis labels
    ax.set_xlim(0, 1)
    ax.set_ylim(-0.5, 0.5)
    ax.axis('off')

    # Adding title
    ax.set_title(title)

    # Showing the plot
    st.pyplot(fig)

# Streamlit app
st.title('Stand Ruler Chart')

# Input values
value = st.slider('Value:', min_value=0, max_value=100, value=50)
max_value = st.number_input('Max Value:', min_value=value, value=100)
title = st.text_input('Chart Title:', 'Stand Ruler Chart')

# Creating and displaying the chart
stand_ruler_chart(value, max_value, title)
```

```python
def uang(angka):
    if angka >= 10**12:
        return f"{angka / 10**12:.0f} T"
    elif angka >= 10**9:
        return f"{angka / 10**9:.0f} M"
    elif angka >= 10**6:
        return f"{angka / 10**6:.0f} Jt"
    elif angka >= 10**3:
        return f"{angka / 10**3:.0f} Rb"
    elif angka >= 10**12:
        return f"{angka / 10**12:.0f} T"
    elif angka <= -10**9:
        return f"{angka / 10**9:.0f} M"
    elif angka <= -10**6:
        return f"{angka / 10**6:.0f} Jt"
    elif angka <= -10**3:
        return f"{angka / 10**3:.0f} Rb"
    else:
        return str(angka)
```

```python
import csv
from modulenorm.modNormalize import normalize
from modulenorm.modTokenizing import tokenize

# fungsi normalisasi
def normalisasi(text):
	text = text.encode("utf-8")
	text_decode = str(text.decode("utf-8"))

	# isisialisasi modul normalisasi dari folder modulnorm/modNormalize
	# isinya normalisasi: lower, repeat char, space char, link, emoticon, ellipsis, spell
	usenorm = normalize()

	# normalisasi enter, 1 revw 1 baris
	text_norm = usenorm.enterNormalize(text_decode)
	# normalisasi spesial char
	text_norm = "".join([x for x in text_norm if ord(x)<128])
	# normalisasi titik yang berulang
	text_norm = usenorm.repeatcharNormalize(text_norm)
	# normalisasi link dalam text
	text_norm = usenorm.linkNormalize(text_norm)
	# normalisasi spasi karakter
	text_norm = usenorm.spacecharNormalize(text_norm)
	# normalisasi karakter elepsis (…)
	text_norm = usenorm.ellipsisNormalize(text_norm)
	# hapus kutip jika ada
	text_norm = text_norm.replace("'", "")

	# panggil modul tokenisasi untuk persiapan tokenisasi
	tok = tokenize()
	# pisah tiap kata pada kalimat
	text_norm = tok.WordTokenize(text_norm)

	# cek spell dari kata perkata
	text_norm = usenorm.spellNormalize(text_norm)
	# menyambung kata (malam-malam) (param: textlist, jmlh_loop)
	text_norm = usenorm.wordcNormalize(text_norm,2)
	# menggabung kalimat tokenize dengan separate spasi
	text_norm = ' '.join(text_norm)
	# menggabung struktur emoticon yang terpisah ([: - )] = [:-)])
	text_norm = usenorm.emoticonNormalize(text_norm)

	# kembalikan hasil normalisasi
	return text_norm

```

```python
# Eksekusi #
# inisialisasi file yang akan dinormalisasi
target = "news.csv";

# buka file csv
with open(target, encoding = "ISO-8859-1") as csvfile:
	# beritahu, kolom dalam csv dipisah dengan pipe (|)
	readCSV = csv.reader(csvfile, delimiter=',')
	# untuk tiap baris data dalam csv lakukan perulangan
	for text in readCSV:

		# ambil text asli
		text_asli = text[0]
		# lakukan normalisasi dari text asli
		text_normalized = normalisasi(text_asli)

		# tampilkan hasil normalisasi text
		print("#####################")
		print("text asli: "+text_asli)
		print("text normalisasi: "+text_normalized)

		# simpan hasil normalisasi dalam file
		output = open("data_normalized.csv","a")
		output.write(str(text_normalized))
		output.write('\n')
		output.close()

	print("#####################")
	print("lihat hasil normalisasi di file data_normalized.csv")

# jika ada error sastrawi: pip install Sastrawi
```

```python
import numpy as np

def calculate_percentiles(data, mean):
    # Menghitung percentil 25, 50 (median), dan 75
    percentile_25 = np.percentile(data, 25)
    percentile_50 = np.percentile(data, 50)  # Median
    percentile_75 = np.percentile(data, 75)

    # Menghitung nilai percentil berdasarkan mean
    percentile_mean = np.percentile(data, np.searchsorted(np.sort(data), mean) / len(data) * 100)

    return percentile_25, percentile_50, percentile_75, percentile_mean

# Contoh penggunaan
data = [10, 15, 20, 25, 30]  # Ganti dengan data yang sesuai
mean_value = np.mean(data)

percentile_25, percentile_50, percentile_75, percentile_mean = calculate_percentiles(data, mean_value)

print(f"Data: {data}")
print(f"Mean: {mean_value}")
print(f"Percentile 25: {percentile_25}")
print(f"Percentile 50 (Median): {percentile_50}")
print(f"Percentile 75: {percentile_75}")
print(f"Percentile Mean: {percentile_mean}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
#import streamlit as st

# Contoh data
data = {
    'p': [1, 2, 3, 4, 5],
    'om': [10, 20, 30, 40, 50],
    'kode': ['A', 'B', 'C', 'D', 'E']
}
s = pd.DataFrame(data)

# Buat scatter plot
fig, ax = plt.subplots()
x = s['p']
y = s['om']
kd = s['kode']
sns.scatterplot(x=x, y=y, marker='>', ax=ax)
plt.xlabel("Posisi Harga")
plt.ylabel("Margin Operasi (%)")

# Anotasi titik data
for a, b, c in zip(x, y, kd):
    label = f"{c} {int(b)}%"
    ax.annotate(label, (a, b), xytext=(3, -3), textcoords='offset points', fontsize=7)

# Tambahkan event handler untuk menampilkan info saat mouse hover
def on_hover(event):
    if event.inaxes == ax:
        for i in range(len(x)):
            if abs(event.xdata - x[i]) < 0.1 and abs(event.ydata - y[i]) < 0.1:
                print("Kode: {kd[i]}, Posisi Harga: {x[i]}, Margin Operasi: {y[i]}%")

fig.canvas.mpl_connect('motion_notify_event', on_hover)

# Tampilkan plot menggunakan st.pyplot()
fig
```
